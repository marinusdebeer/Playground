```python
import ast
import numpy as np
import matplotlib.pyplot as plt

def plot_multiple_runs():
    # folders = ["per_30_5_000_g", "per_31_5_000_g", "per_34_5_000_i", "per_35_5_000_i", "per_37_5_000_k", "per_38_5_000_k"]
    folders = ["per_34_5_000_i", "per_37_5_000_k", "per_38_5_000_k", "per_39_5_000_l", "per_40_5_000_l"]
    data = {}
    for folder in folders:
        with open(f"G:/Coding/breakout/testing_prioritized/{folder}/rewards.txt", "r") as f:
            data_str = f.read()
            data[folder] = ast.literal_eval(data_str)
    
    # Create a single plot with a larger figsize
    fig, ax = plt.subplots(figsize=(20, 6))
    
    for folder in folders:
        data_array = np.array(data[folder])
        total = 1500
        interval = 100
        reshaped_arr = data_array[:total].reshape(total // interval, interval)
        averaged_arr = reshaped_arr.mean(axis=1)
        
        print(folder)
        print(averaged_arr)
        print()
        
        # Plot on the single axis, ax
        ax.plot(averaged_arr, label=folder)
    
    ax.set_title('Comparison of Different Runs')
    ax.set_ylabel('Rewards')
    ax.set_xlabel('Episodes')
    ax.set_ylim([0, 30])  # Set y-axis limits
    ax.legend()  # Show legend to identify each folder
    ax.grid(True)
    plt.show()

# plot_multiple_runs()
```
