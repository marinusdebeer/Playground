```python
import matplotlib.pyplot as plt
import numpy as np

def epsilon_decay(initial_epsilon, decay_rate, episode, num_episodes, minimum):
    epsilon = max((initial_epsilon * (decay_rate ** (episode/num_episodes))), minimum)
    return epsilon

initial_epsilon = 1.0  # Starting epsilon value
decay_rate = 0.04  # Decay rate
num_episodes = 15_000  # Total number of episodes

# Pre-allocate lists to hold episode numbers and corresponding epsilon values
episodes = np.arange(0, num_episodes+1)
eps_values = []

# Calculate epsilon for each episode and store it
for ep in episodes:
    if ep > 10_000:
        eps_values.append(epsilon_decay(0.10, decay_rate, ep - 10_000, 3000, 0.01))
    elif ep > 6_000:
        eps_values.append(epsilon_decay(0.20, decay_rate, ep - 6_000, 4000, 0.0))
    elif ep > 3_000:
        eps_values.append(epsilon_decay(0.25, decay_rate, ep - 3_000, 2000, 0.01))
    else:
        eps_values.append(epsilon_decay(initial_epsilon, decay_rate, ep, 2000, 0.03))

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(episodes, eps_values)
plt.xlabel('Episodes')
plt.ylabel('Epsilon')
plt.title('Epsilon Decay over Episodes')
plt.grid(True)
plt.show()
```
